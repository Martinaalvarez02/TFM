```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

# Definir la matriz de confusión
cm = [[93, 2],
      [4, 5]]

# Definir los nombres de las clases
clases = ['LIHC', 'CHOL']

# Crear la matriz de confusión como DataFrame
df_cm = pd.DataFrame(cm, columns=clases, index=clases)
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'

# Crear figura
plt.figure(figsize=(10, 7))
sns.set_theme(font_scale=2.5)

# Crear mapa de calor
heatmap = sns.heatmap(df_cm, cmap="Blues", annot=False, cbar=False)
heatmap.set_aspect("equal")

# Agregar anotaciones manuales
for i in range(len(clases)):
    for j in range(len(clases)):
        text = '{:.0f}'.format(cm[i][j])
        color = 'black' if cm[i][j] < 128 else 'white'
        plt.text(j + 0.5, i + 0.5, text, ha='center', va='center', color=color, fontsize=45)

# Guardar la imagen antes de mostrarla
name = 'matriz_confusion_5_cambiada.png'
plt.savefig(name, bbox_inches='tight')

# Mostrar la figura
plt.show()

# Mostrar la imagen guardada directamente en Colab
Image.open(name).show()
```
